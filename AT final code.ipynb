```python
# -------------------- IMPORTS --------------------


import os
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.linear_model import BayesianRidge
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, GRU, Dense, Dropout, Bidirectional
import tensorflow as tf
from datetime import datetime, timedelta

# -------------------- SEED SETTING --------------------
def set_seed(seed=42):
    os.environ['PYTHONHASHSEED'] = str(seed)
    random.seed(seed)
    np.random.seed(seed)
    tf.random.set_seed(seed)

set_seed(42)

# -------------------- DATA FETCHING --------------------
today_date = datetime.today().date()
yesterday_date = today_date - timedelta(days=1)
start_date = '2014-01-01'

df = yf.download('WMT', start=start_date, end=yesterday_date.strftime('%Y-%m-%d'))
data = df['Close'].values.reshape(-1, 1)
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data)

train_size = int(len(scaled_data) * 0.8)
train_data, test_data = scaled_data[:train_size], scaled_data[train_size:]

# -------------------- DATA PREPARATION --------------------
def create_dataset(dataset, look_back=60):
    X, Y = [], []
    for i in range(len(dataset) - look_back):
        X.append(dataset[i:(i + look_back), 0])
        Y.append(dataset[i + look_back, 0])
    return np.array(X), np.array(Y)

look_back = 60
X_train, y_train = create_dataset(train_data, look_back)
X_test, y_test = create_dataset(test_data, look_back)

X_train_rnn = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
X_test_rnn = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)

# -------------------- EVALUATION FUNCTION --------------------
def evaluate_model(name, y_true, y_pred):
    y_true = y_true.reshape(-1, 1)
    y_pred = y_pred.reshape(-1, 1)
    y_true_inv = scaler.inverse_transform(y_true)
    y_pred_inv = scaler.inverse_transform(y_pred)

    mse = mean_squared_error(y_true_inv, y_pred_inv)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_true_inv, y_pred_inv)
    r2 = r2_score(y_true_inv, y_pred_inv)
    accuracy = max(0, (1 - (rmse / np.mean(y_true_inv))) * 100)

    print(f"\n{name} Performance:")
    print(f"Accuracy: {accuracy:.2f}%")
    print(f"RMSE: {rmse:.3f}")
    print(f"MAE: {mae:.3f}")
    print(f"MSE: {mse:.3f}")
    print(f"R Score: {r2:.3f}")

    return y_pred_inv, y_true_inv, rmse, mse, r2

# -------------------- LSTM MODEL --------------------
lstm_model = Sequential([
    LSTM(50, return_sequences=True, input_shape=(look_back, 1)),
    Dropout(0.3),
    LSTM(50),
    Dropout(0.3),
    Dense(1)
])
lstm_model.compile(optimizer='adam', loss='mean_squared_error')
lstm_model.fit(X_train_rnn, y_train, epochs=20, batch_size=32, validation_data=(X_test_rnn, y_test), verbose=0)

lstm_test_pred = lstm_model.predict(X_test_rnn)
lstm_pred_inv, y_test_inv, lstm_rmse, lstm_mse, lstm_r2 = evaluate_model("LSTM", y_test, lstm_test_pred)

# -------------------- GRU MODEL --------------------
gru_model = Sequential([
    GRU(50, return_sequences=True, input_shape=(look_back, 1)),
    Dropout(0.3),
    GRU(50),
    Dropout(0.3),
    Dense(1)
])
gru_model.compile(optimizer='adam', loss='mean_squared_error')
gru_model.fit(X_train_rnn, y_train, epochs=20, batch_size=32, validation_data=(X_test_rnn, y_test), verbose=0)

gru_test_pred = gru_model.predict(X_test_rnn)
gru_pred_inv, _, gru_rmse, gru_mse, gru_r2 = evaluate_model("GRU", y_test, gru_test_pred)

# -------------------- BiLSTM MODEL --------------------
bilstm_model = Sequential([
    Bidirectional(LSTM(50, return_sequences=True), input_shape=(look_back, 1)),
    Dropout(0.3),
    Bidirectional(LSTM(50)),
    Dropout(0.3),
    Dense(1)
])
bilstm_model.compile(optimizer='adam', loss='mean_squared_error')
bilstm_model.fit(X_train_rnn, y_train, epochs=20, batch_size=32, validation_data=(X_test_rnn, y_test), verbose=0)

bilstm_test_pred = bilstm_model.predict(X_test_rnn)
bilstm_pred_inv, _, bilstm_rmse, bilstm_mse, bilstm_r2 = evaluate_model("BiLSTM", y_test, bilstm_test_pred)

# -------------------- BAYESIAN RIDGE --------------------
X_train_flat = X_train
X_test_flat = X_test
bayesian_model = BayesianRidge()
bayesian_model.fit(X_train_flat, y_train)

bayes_test_pred = bayesian_model.predict(X_test_flat)
bayes_pred_inv, _, bayes_rmse, bayes_mse, bayes_r2 = evaluate_model("Bayesian Ridge", y_test, bayes_test_pred)

# -------------------- OVERFITTING CHECK --------------------
def check_overfit(name, y_train_pred, y_test_pred, y_train_true, y_test_true):
    y_train_pred = y_train_pred.reshape(-1, 1)
    y_test_pred = y_test_pred.reshape(-1, 1)
    y_train_true = y_train_true.reshape(-1, 1)
    y_test_true = y_test_true.reshape(-1, 1)

    train_pred_inv = scaler.inverse_transform(y_train_pred)
    test_pred_inv = scaler.inverse_transform(y_test_pred)
    train_true_inv = scaler.inverse_transform(y_train_true)
    test_true_inv = scaler.inverse_transform(y_test_true)

    rmse_train = np.sqrt(mean_squared_error(train_true_inv, train_pred_inv))
    rmse_test = np.sqrt(mean_squared_error(test_true_inv, test_pred_inv))
    print(f"{name} Overfit Check -> Train RMSE: {rmse_train:.3f}, Test RMSE: {rmse_test:.3f}, Gap: {abs(rmse_train - rmse_test):.3f}")

print("\n--- Overfitting Check ---")
check_overfit("LSTM", lstm_model.predict(X_train_rnn), lstm_test_pred, y_train, y_test)
check_overfit("GRU", gru_model.predict(X_train_rnn), gru_test_pred, y_train, y_test)
check_overfit("BiLSTM", bilstm_model.predict(X_train_rnn), bilstm_test_pred, y_train, y_test)
check_overfit("Bayesian Ridge", bayesian_model.predict(X_train_flat), bayes_test_pred, y_train, y_test)

# -------------------- PLOT PREDICTIONS --------------------
plt.figure(figsize=(14,6))
plt.plot(y_test_inv, label='Actual Price', color='blue')
plt.plot(lstm_pred_inv, label='LSTM Prediction', color='red', linestyle='--')
plt.plot(gru_pred_inv, label='GRU Prediction', color='green', linestyle='-.')
plt.plot(bilstm_pred_inv, label='BiLSTM Prediction', color='purple', linestyle='dashdot')
plt.plot(bayes_pred_inv, label='Bayesian Ridge Prediction', color='orange', linestyle=':')
plt.title('Walmart (WMT) Stock Price Prediction')
plt.xlabel('Time')
plt.ylabel('Stock Price')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# -------------------- BAR CHART COMPARISON --------------------
model_names = ["LSTM", "GRU", "BiLSTM", "Bayesian Ridge"]
rmse_scores = [lstm_rmse, gru_rmse, bilstm_rmse, bayes_rmse]
mse_scores = [lstm_mse, gru_mse, bilstm_mse, bayes_mse]
r2_scores = [lstm_r2, gru_r2, bilstm_r2, bayes_r2]

x = np.arange(len(model_names))
width = 0.2

# RMSE
plt.figure(figsize=(12, 6))
plt.bar(x - width, rmse_scores, width, label='RMSE', color='skyblue')
plt.xticks(x, model_names)
plt.ylabel('RMSE Value')
plt.title('RMSE Comparison of Models')
plt.legend()
plt.tight_layout()
plt.show()

# MSE
plt.figure(figsize=(12, 6))
plt.bar(x - width, mse_scores, width, label='MSE', color='salmon')
plt.xticks(x, model_names)
plt.ylabel('MSE Value')
plt.title('MSE Comparison of Models')
plt.legend()
plt.tight_layout()
plt.show()

# R
plt.figure(figsize=(12, 6))
plt.bar(x - width, r2_scores, width, label='R Score', color='lightgreen')
plt.xticks(x, model_names)
plt.ylabel('R Score')
plt.title('R Score Comparison of Models')
plt.legend()
plt.tight_layout()
plt.show()

# -------------------- TODAY'S PREDICTION --------------------
print("\n--- Predicting Today's Closing Price ---")
updated_df = yf.download('WMT', start='2014-01-01', end=today_date.strftime('%Y-%m-%d'))
full_scaled_data = scaler.fit_transform(updated_df['Close'].values.reshape(-1, 1))
latest_sequence = full_scaled_data[-look_back:]
latest_sequence_rnn = latest_sequence.reshape(1, look_back, 1)
latest_sequence_flat = latest_sequence.reshape(1, -1)

lstm_today = scaler.inverse_transform(lstm_model.predict(latest_sequence_rnn))[0][0]
gru_today = scaler.inverse_transform(gru_model.predict(latest_sequence_rnn))[0][0]
bilstm_today = scaler.inverse_transform(bilstm_model.predict(latest_sequence_rnn))[0][0]
bayesian_today = scaler.inverse_transform(bayesian_model.predict(latest_sequence_flat).reshape(-1, 1))[0][0]

print(f"\nPredicted Walmart (WMT) Closing Price for {today_date.strftime('%Y-%m-%d')}:")
print(f"LSTM: ${lstm_today:.2f}")
print(f"GRU: ${gru_today:.2f}")
print(f"BiLSTM: ${bilstm_today:.2f}")
print(f"Bayesian Ridge: ${bayesian_today:.2f}")
```
